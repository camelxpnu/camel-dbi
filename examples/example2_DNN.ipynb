```python
import tensorflow as tf
import numpy as np
from tensorflow import keras
from keras.layers import Dense, Activation, InputLayer, BatchNormalization, BatchNormalization
from keras.models import Sequential, load_model
from sklearn.model_selection import train_test_split
```

```python
## 1. Load Data sets & Data pre-processing

outfile = np.load('Enter your data file NAME and $PATH')
print(outfile.files)               
input_data = outfile['Input Data'] 
output_data = outfile['Output Data']

# Reshape Input data to use in DNN(Deep Neural Network)
input_data = input_data.T
print(np.shape(input_data))
print(np.shape(output_data))

# Split Data sets into training sets, validation sets and test sets
Train_size = 0.2
Test_size = 0.5
X_train, X_test, y_train, y_test = train_test_split(input_data, output_data, random_state=66, test_size=1-Train_size)
X_test, X_val, y_test, y_val = train_test_split(X_test, y_test, random_state=66, test_size=0.5)

print(np.shape(X_train))
print(np.shape(y_train))
print(np.shape(X_test))
print(np.shape(y_test))
print(np.shape(X_val))
print(np.shape(y_val))

# Now, the ratio of training sets, validation sets and test sets is 8 to 1 to 1. 
# (num. of train. : num. of valid. : num. of test = 8 : 1 : 1)

```

```python
## 2. Construct Neural Network architecture : Deep Neural Networks

model = Sequential()

model.add(Dense(30))
model.add(BatchNormalization())
model.add(Activation('swish'))

model.add(Dense(30))
model.add(BatchNormalization())
model.add(Activation('swish'))

model.add(Dense(30))
model.add(BatchNormalization())
model.add(Activation('swish'))

model.add(Dense(1))

model.compile(optimizer='Adam',loss='mse',metrics=['mse'])
```

```python
## 3. Start learning!
batch_size = 128
model.fit(X_train, y_train, batch_size=batch_size,epochs=1,validation_data=(X_val, y_val))

scores = model.evaluate(X_test, y_test)
print('Results of TEST set : ', scores)

model.summary()
```
